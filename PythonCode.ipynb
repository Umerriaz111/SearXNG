```python
import requests

def search_searxng_with_headers(query):
    searxng_url = 'http://127.0.0.1:8080/search'
    headers = {
        'User -Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    
    params = {
        'q': query,
        'format': 'json',
        'category': 'general',
    }
    
    try:
        response = requests.get(searxng_url, headers=headers, params=params)
        response.raise_for_status()
        results = response.json()
        return results
    
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
if __name__ == "__main__":
    query = "Shoes Under $10"
    results = search_searxng_with_headers(query)
    
    if results:
        print("Search Results:")
        for result in results.get('results', []):
            print(f"- {result['title']}: {result['url']}")
```

```python
import requests

def search_searxng_get(query):
    searxng_url = 'http://127.0.0.1:8080/search'
 
    params = {
        'q': query,
        'format': 'json',
        'category': 'general',  # Optional: specify categories if needed
    }
    
    try:
        response = requests.get(searxng_url, params=params)
        response.raise_for_status()
        results = response.json()
        return results
    
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
if __name__ == "__main__":
    query = "Python programming"
    
    # Using GET method
    print("Using GET method:")
    results_get = search_searxng_get(query)
    if results_get:
        for result in results_get.get('results', []):
            print(f"- {result['title']}: {result['url']}")
    
    
```

# 2 POST Request

```python
def search_searxng_post(query):
    searxng_url = 'http://127.0.0.1:8080/search'
   
    
    data = {
        'q': query,
        'format': 'json',
        'category': 'general',  # Optional: specify categories if needed
    }
    
    try:
        response = requests.post(searxng_url, data=data)
        response.raise_for_status()
        results = response.json()
        return results
    
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
```

```python
# Using POST method
print("\nUsing POST method:")
query="Shoes under $10 dollar for Childrens"
results_post = search_searxng_post(query)
if results_post:

    for result in results_post.get('results', []):
        print("************************************")
        print("Result is From :",result['engine'])
        print("Title is : ",result['title'])
        print('Content is : ',result['content'])
        print("Url is :",result['url'])
        print("************************************")
                  
                  
```

```python
import requests
from io import BytesIO
from PIL import Image
import matplotlib.pyplot as plt

def plot_image_from_url(image_url):
    # Fetch the image
    response = requests.get(image_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Open the image using PIL
        img = Image.open(BytesIO(response.content))
        
        # Plot the image
        plt.figure(figsize=(3, 3))
        plt.imshow(img)
        plt.axis('off')  # Hide axis for a cleaner look
        plt.show()
    else:
        print(f"Error: Unable to fetch image. Status code: {response.status_code}")

```

```python
import requests

def search_searxng_get(query, num_results=10):
    searxng_url = 'http://127.0.0.1:8080/search'
    
    data = {
        'q': query,
        'format': 'json',
        'category': 'images',  # Optional: specify categories if needed
        'engines': 'bing images',
        'num': num_results  # Assuming the API has a 'num' parameter to limit results
    }
    
    try:
        response = requests.get(searxng_url, params=data)  # Use 'params' instead of 'data'
        response.raise_for_status()
        results = response.json()
        
        # Ensure 'images' are available in the response
        if 'images' in results:
            image_results = results['images']
            total_results = len(image_results)
            print(f"Total results found (limited to {num_results}): {total_results}")
            return image_results
        else:
            print("No image results found.")
            return None
    
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None

# Example usage:
query = "shoes under 10 dollars for women"
results = search_searxng_get(query, num_results=10)

# Display the results (for example, display the image URLs)
if results:
    for i, result in enumerate(results):
        print(f"Result {i+1}: {result['url']}")
```
